```python
df['start_time'] = pd.to_datetime(df['start_time'])
df['end_time'] = pd.to_datetime(df['end_time'])
df['time'] = df['end_time'] - df['start_time']
df['time'] = df['time'].apply(lambda x: str(x).split()[-1])  # Extract HH:MM:SS part
```

```python
grouped_workouts = df.groupby(['workout_id', 'time']).agg(
    num_exercises=('excercise_title', 'nunique'),  # Count unique exercises
)
```

```python
grouped_exercises = df.groupby(['workout_id','excercise_title', 'start_time', 'end_time']).agg(
    num_of_sets=('set_index', 'count')  # Average sets per workout
)
```
